Returning using the data extracting from CoinMarket

Extracting the table from the website

```python
#Extracting the table for a pandas df
data = CoinMarket.find_all('tr')
```

```python
data_without_header = data[1:]
headers = data[0]
```

```python
col_headers = [th.getText() for th in headers.findAll('th')]
df_data = [[td.getText() for td in tr.findAll('td')] for tr in data_without_header]
```

```python
df = pd.DataFrame(df_data, columns = col_headers)
df.head()
```

```python
#Creating a new DataFrame with just the 5 currencies on top (highest Volume)
top5 = df.head(5)
```

```python
#Removing special characteres ($ , %) and letters
top5['Price'] = top5['Price'].map(lambda x: re.sub(r'\$+', '', x))
top5['Price'] = top5['Price'].map(lambda x: re.sub(r'\,+', '', x))
top5['24h %'] = top5['24h %'].map(lambda x: re.sub(r'\%+', '', x))
top5['7d %'] = top5['7d %'].map(lambda x: re.sub(r'\%+', '', x))
top5['Market Cap'] = top5['Market Cap'].map(lambda x: re.sub(r'\$+', '', x))
top5['Market Cap'] = top5['Market Cap'].map(lambda x: re.sub(r'\,', '', x))
top5['Market Cap'] = top5['Market Cap'].map(lambda x: re.sub(r'[^0-9]', '', x))
top5['Volume(24h)'] = top5['Volume(24h)'].map(lambda x: re.sub(r'\$+', '', x))
top5['Volume(24h)'] = top5['Volume(24h)'].map(lambda x: re.sub(r'\,', '', x))
top5['Volume(24h)'] = top5['Volume(24h)'].map(lambda x: re.sub(r'[^0-9]', '', x))
top5['Circulating Supply'] = top5['Circulating Supply'].map(lambda x: re.sub(r'\,', '', x))
top5['Circulating Supply'] = top5['Circulating Supply'].map(lambda x: re.sub(r'[^0-9]', '', x))
top5
```

## Plots

```python
#Seting the numbers as float
top5['Price'] = top5['Price'].astype(float, errors = 'raise')
top5['24h %'] = top5['24h %'].astype(float, errors = 'raise')
top5['7d %'] = top5['7d %'].astype(float, errors = 'raise')
top5['Market Cap'] = top5['Market Cap'].astype(float, errors = 'raise')
top5['Volume(24h)'] = top5['Volume(24h)'].astype(float, errors = 'raise')
top5['Circulating Supply'] = top5['Circulating Supply'].astype(float, errors = 'raise')
```

When purchasing one unit of each of the five currencies, Bitcoin alone will account for almost the entire purchase value. The chart below clearly shows that Bitcoin has the highest percentage value among the five coins. Ethereum is the only one that comes close in terms of value. As such, Bitcoin is significantly more expensive than the other four coins combined.

```python
plot = top5.plot.pie(y='Price', figsize=(8, 8))
```

The percentage "24%" reflects the trading volume in the last 24 hours, which is a crucial factor for investors to consider as it determines the investment's liquidity - how easily it can be converted to cash. Poorly liquid investments can make it challenging to sell assets at a favorable price in the short term, even if the asset increases in value.

According to the data, Tether and USDCoin showed lower liquidity compared to other investments, making them relatively more difficult to convert into dollars.

```python
top5.plot(x="Name", y=["24h %"], kind="bar") 
```

The chart below depicts the circulation density of the top 5 cryptocurrencies, and it reveals significant variations. This highlights the fact that the cryptocurrency market is not as stable as that of traditional currencies.

```python
top5['Circulating Supply'].plot.kde(bw_method=0.3)
```

The scatter plot depicted below showcases the coins based on their price and total circulation volume (both in absolute and percentage terms). Once again, it becomes evident that the values exhibit a considerable level of variation.

```python
top5.plot.scatter(x='Price',
                  y='24h %',
                  c='Volume(24h)',
                  colormap='viridis')
```

The Market Cap of a cryptocurrency is determined by multiplying its price with the circulating supply, and it indicates the total market value of the cryptocurrency's circulating supply.

The chart below reveals that the combined Market Cap of Bitcoin and Ethereum makes up a vast majority of the total value of cryptocurrencies in circulation.

```python
top5.plot.area(x="Name",y='Market Cap', figsize=(8, 8))
```
